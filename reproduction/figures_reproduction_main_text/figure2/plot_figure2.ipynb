```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

metrics = ['1 Neuron','2 Neurons','3 Neurons','4 Neurons','5 Neurons']
results_AUROC = {
    'SegNeuro-2D':  [0.5357,0.9829,1.2258,1.4093,1.6298],
    'SAM':  [0.5381,1.1604,1.5825,1.9891,2.2882],
}
```

```python
import matplotlib.pyplot as plt
import numpy as np

color_dict = {
    'SegNeuro-2D':   'r',
    'SAM':           'k',
}

# 创建示例数据
labels = np.array(metrics)
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for method in results_AUROC.keys():
    values = np.array(results_AUROC[method])

    # 绘制雷达图
    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    values = np.concatenate((values,[values[0]]))
    angles += angles[:1]
    
    # ax.fill(angles, values, color='none', edgecolor=color_dict[method], alpha=1, linestyle='-')
    ax.fill(angles, values, color='none', edgecolor=color_dict[method], alpha=1, linestyle='-')

    # 添加标签
    # ax.set_ylim([0,1])
    # ax.set_yticklabels([0.2,0.4,0.6,0.8,1])
    # ax.set_ylim([0.75,1])
    # ax.set_yticks([0.75,0.8,0.85,0.90,0.95,1])
    # ax.set_yticklabels([0.8,0.85,0.90,0.95,1])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=12)

plt.legend(results_AUROC.keys())
plt.savefig('./figure2b.pdf', dpi=300)  # eps文件，用于LaTeX
```
