```python
voi_dict_SegNeuro2D_star = np.load('./voi_dict_SegNeuro2D(star).npy',allow_pickle=True).item()
voi_dict_SegNeuro2D = np.load('./voi_dict_SegNeuro2D.npy',allow_pickle=True).item()
voi_dict_ACRLSD2D = np.load('./voi_dict_ACRLSD-2D.npy',allow_pickle=True).item()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np


# 输入统计数据
# xstick_names = list(voi_dict_SegNeuro2D_star.keys())
xstick_names = np.arange(1,len(voi_dict_SegNeuro2D_star)+1)
value_1 = [np.mean(np.sum(voi_dict_SegNeuro2D_star[key],axis=1)) for key in voi_dict_SegNeuro2D_star.keys()]
value_2 = [np.mean(np.sum(voi_dict_ACRLSD2D[key],axis=1)) for key in voi_dict_ACRLSD2D.keys()]
value_3 = [np.mean(np.sum(voi_dict_SegNeuro2D[key],axis=1)) for key in voi_dict_SegNeuro2D.keys()]

bar_width = 0.2  # 条形宽度
index_value_1 = np.arange(len(xstick_names))  
index_value_2 = index_value_1 + bar_width  
index_value_3 = index_value_2 + bar_width  

# 使用两次 bar 函数画出两组条形图
plt.figure(figsize=(20,4))
plt.bar(index_value_1, height=value_1, width=bar_width, color='#2E7EBB', label='UniSPAC-2D')
plt.bar(index_value_2, height=value_2, width=bar_width, color='#E25508', label='ACRLSD-2D')
plt.bar(index_value_3, height=value_3, width=bar_width, color='#5A5A5A', label='UniSPAC-2D*')

plt.legend()  # 显示图例
plt.xticks(index_value_1 + bar_width/2, xstick_names)  # 让横坐标轴刻度显示 xstick_names 里的饮用水， index_value_1 + bar_width/2 为横坐标轴刻度的位置
plt.ylabel('VOI')  # 纵坐标轴标题
plt.title('results')  # 图形标题
plt.xlabel("Volumes number")

# plt.show()
plt.savefig('./figure4_b.pdf', dpi=300) 
```
